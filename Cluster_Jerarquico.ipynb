```python
# Hierarchical Clustering visual con Dendrograma
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
# Configuración de estilo visual
plt.style.use('ggplot') # Using a standard matplotlib style
plt.rcParams['figure.figsize'] = (10, 7) # Set default figure size
plt.rcParams['font.size'] = 12 # Set default font size
plt.rcParams['axes.labelsize'] = 12 # Set default axes label size
plt.rcParams['axes.titlesize'] = 14 # Set default axes title size
plt.rcParams['xtick.labelsize'] = 10 # Set default x-tick label size
plt.rcParams['ytick.labelsize'] = 10 # Set default y-tick label size
```

```python
# Generar datos
X, _ = make_blobs(n_samples=60, centers=3, cluster_std=2.9, random_state=42)
```

```python
# Visualizar dataset inicial
plt.figure(figsize=(7,5))
plt.scatter(X[:,0], X[:,1], s=80, c='gray', alpha=0.6)
plt.title("Datos iniciales sin clasificar")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()
```

```python
# Aplicar linkage (método de Ward)
Z = linkage(X, method='ward')

# Dibujar dendrograma
plt.figure(figsize=(6, 4))
dendrogram(Z, color_threshold=10, above_threshold_color='gray')
plt.title("Clustering Jerárquico (Dendrograma)")
plt.xlabel("Muestras")
plt.ylabel("Distancia")
plt.tight_layout()
plt.show()
```
